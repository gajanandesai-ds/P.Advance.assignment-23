```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0,math.pi,0.01)
y = np.sin(20*x)

plt.figure(figsize = (6,5))
plt.plot(x,y)
plt.title('plot with undefined aspect ratio')
plt.show()
```

```python
plt.figure(figsize = (10,3))
plt.plot(x,y)
plt.title('plot with undefined aspect ratio')
plt.show()
```

```python
plt.figure(figsize=(10,3))
ax = plt.gca()
ax.set_aspect(1.5)


plt.plot(x,y)
plt.title('plot with fixed aspect rat')
plt.show()
```

```python
import numpy as np

v1 = np.array([[1,2],[1,2]])

v2 = np.array([[1,2],[1,2]])

print("vector multiplication")
print(np.dot(v1,v2))
print("outer multiplication")
print(np.outer(v1,v2))
print("\nElementwise multiplication of two vector")
print(v1*v2)
```
